```python
import pandas as pd
import numpy as np
```

### Ratings where there are clear themes, fighting vs non fighting


```python
df = pd.read_csv('arcade.csv')
df.drop(df.columns[0], axis=1, inplace=True)
df.drop(df.columns[-1], axis=1, inplace=True)
df.drop(11, inplace=True)
```

```python
from scipy.sparse.linalg import svds
from scipy import dot
df_matrix = df.as_matrix()
#df_matrix.asfp
df_matrix = df_matrix.astype(float)
```

### Try k=2

```python
U, sigma1, Vt = svds(df_matrix, k = 2)
sigma = np.diag(sigma1)
print 'here is our user:concept matrix'
print U
print 'concept:concept matrix'
print sigma
print 'game : concept matrix'
print Vt

```

### When we set k=2, can see there are 2 clear camps. Fighting vs non fighting. 

### The U and Vt matrix had literally no overlap.

```python

def create_tester(dataframe, k): 
    '''Given our ratings dataframe, select the first k rows, and transform 50% of the values to 0. We now have a 
    ratings dataframe with some hidden values, and we can perform recommendations on it, to compare with our,
    ground truth ratings dataframe.
    Arguments
    dataframe: input your ratings dataframe, where rows are users, columns are items.
    k: number of rows to slice off. These rows will then have some of their values hidden.
    '''
    
    #Create array of the first K rows
    temp_array = np.array(dataframe.iloc[0:k])
    #Change 50% of the values to 0.
    for index in np.nditer(temp_array, op_flags=['readwrite']):
        chance = np.random.randint(1,3) 
        if chance == 1:
            index[...] = 0 
    
    #Return a dataframe that has been modified by 'hiding' ratings, now ready to perform testing. 
    temp_df = pd.DataFrame(temp_array)
    result = temp_df.append(dataframe.drop(dataframe.index[:k], axis=0))
    return result
```

```python
tester_matrix = create_tester(df, 5)
```

```python
tester_matrix
```

```python
tester_matrix
```

```python
#Perform svd on tester_matrix

df_matrix = tester_matrix.as_matrix()
#df_matrix.asfp
df_matrix = df_matrix.astype(float)

U, sigma1, Vt = svds(df_matrix, k = 2)
sigma = np.diag(sigma1)
print 'here is our user:concept matrix'
print U
print 'concept:concept matrix'
print sigma
print 'game : concept matrix'
print Vt

dotUsigma = np.dot(U, sigma)
preds = np.dot(dotUsigma, Vt)
preds
```

```python
print preds
print df
print preds - np.array(df) #This is the difference. 
```

```python
def rmse(test_df, prediction):
    np.sqrt(sum(sum((prediction - np.array(test_df)) ** 2)) / (preds.shape[0] * preds.shape[1]))
```
