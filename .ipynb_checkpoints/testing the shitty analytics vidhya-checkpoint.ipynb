```python
def striptitle(title):
    charremove = ["(",")","-"]
    for i in charremove:
        if i in title:
            while i in title:
                title = title.replace(i,"")
    return title
```

### Basic prediction generator, using scipy Sparse SVD

```python
# Function to normalise ratings and create predictions matrix via SVD.
# Function to normalise ratings and create predictions matrix via SVD.

def create_svd_predictions(ratings, svd_k, demean=1, return_components=0):
    '''
    Performs SVD on a sparse matrix, where rows are users, columns are items, and each value represents the users rating.
    U is the users to concept matrix. sigma is the concept importance matrix. Vt is the products to concept marix.
    Returns: 
    the matrix of predicted ratings in a pandas DF. return_components=1 will make the function return U, sigma, Vt.
    Args:
    ratings = ratings dataframe where 
    demean = 1 to de-mean by product.
    demean = 2 to de-mean by users.
    
    Example:
    my_predictions, tuple_to_unpack = create_svd_predictions(ratings_dataframe, 10, demean=1, return_components=1)
    U, sigma, Vt = tuple_to_unpack
    ========
    
    '''
    if demean == 1:
        
        # Get mean of non-zero items, and minus off from mean of each non zero rating.
        y = np.ma.masked_where(ratings == 0, ratings)
        product_mean = np.mean(y, axis=1)
        r_demeaned = ratings - product_mean.reshape(-1,1)
        #SVD performed on de-meaned ratings.
        U, sigma1, Vt = svds(r_demeaned, k=svd_k)
        sigma = np.diag(sigma1)
        all_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + product_mean.reshape(-1,1)
        preds_df = pd.DataFrame(all_predicted_ratings, columns=r_df.columns)
        
        if return_components == 1:
            return preds_df, U, sigma1, Vt
        
        return preds_df
    
        
    elif demean==2:
        y = np.ma.masked_where(ratings == 0, ratings)
        user_means = np.mean(y, axis=0)
        r_demeaned = ratings - user_means
    
        U, sigma1, Vt = svds(r_demeaned, k=svd_k)
        sigma = np.diag(sigma1)
        all_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_means
        preds_df = pd.DataFrame(all_predicted_ratings, columns=r_df.columns)
            
        if return_components == 1:
            return preds_df, U, sigma1, Vt


        
        return preds_df

preds_df = create_svd_predictions(r, 20, demean=2)
    
```

#  Top K recommendations generator 

```python
def recommend_movies(predictions_df, userID, movies_df, original_ratings_df, num_recommendations=5):
    '''Generates top k predictions for a user. 
    Args:
    predictions_df: Dataframe of predicted ratings where columns = products, rows = users.
    userID: ID of the user to generate recommendations for.
    movies_df: Dataframe of movies, where movie ID column is named "MovieID"
    original_ratings_df: Original dataframe of matrix where columns = products, rows = users
    '''
    # Get and sort the user's predictions
    user_row_number = userID - 1 # UserID starts at 1, not 0
    sorted_user_predictions = predictions_df.iloc[user_row_number].sort_values(ascending=False)
    
    # Get the user's data and merge in the movie information.
    user_data = original_ratings_df[original_ratings_df.UserID == (userID)]
    user_full = (user_data.merge(movies_df, how = 'left', left_on = 'MovieID', right_on = 'MovieID').
                     sort_values(['Rating'], ascending=False)
                 )

    print 'User {0} has already rated {1} movies.'.format(userID, user_full.shape[0])
    print 'Recommending the highest {0} predicted ratings movies not already rated.'.format(num_recommendations)
    
    # Recommend the highest predicted rating movies that the user hasn't seen yet.
    recommendations = (movies_df[~movies_df['MovieID'].isin(user_full['MovieID'])].
         merge(pd.DataFrame(sorted_user_predictions).reset_index(), how = 'left',
               left_on = 'MovieID',
               right_on = 'MovieID').
         rename(columns = {user_row_number: 'Predictions'}).
         sort_values('Predictions', ascending = False).
                       iloc[:num_recommendations, :-1]
                      ) 
    return user_full, recommendations

already_rated, predictions = recommend_movies(preds_df, 600, movies_df, ratings_df, 10)
```

Process: Hide n% of values from the original ratings matrix, and run algorithm on the matrix. Compare the predicted values.




```python
# Function that creates a matrix with hidden values, and the locations of hidden values.
def create_test_matrix(ground_truth, percent_to_hide):
    '''returns a matrix, with a percentage of non-zero values masked. (aka test set)
    
    Args:
    ground_truth: must be a matrix where rows are users, columns are products, and 
    values are the ratings given.
    percent_to_hide: must be  > 0 , < 1
    
    Creates train test split '''
    
    
    temp_array = np.array(ground_truth)
    total_non_zero = np.count_nonzero(temp_array)
    print 'There are ' + str(total_non_zero) + ' non-zero items.'
    idx = np.random.choice(total_non_zero, int(total_non_zero * percent_to_hide), replace=False)
    
    print 'Hiding ' + str(len(idx)) + ' items.'
    
    temp_array[tuple(map(lambda x: x[idx], np.where(temp_array)))] = 0.0
    small_minus_big = temp_array - ground_truth
    i,j = np.where(small_minus_big != 0)
    locators = zip(i,j)
    return temp_array, locators
```

```python
def get_rmse(ground_truth, predictions, locators):
    # Create hidden ratings dataset, feeding in our ground truth df. 

# differences_list is predictions[locator] - ground_truth[locator]
    differences_list = []
    my_preds_list = []
    ground_truth_list = []
    movie_id_list = []
    
    #turn predictions into numpy array
    predictions_array = np.array(predictions)
    #turn base ratings into numpy array
    ground_truth_array = np.array(ground_truth)

    for i in locators:
        my_preds_list.append(predictions_array[i])
        ground_truth_list.append(ground_truth_array[i])
        
    differences_array = np.array(my_preds_list) - np.array(ground_truth_list)
    rmse = np.sqrt(np.mean(differences_array ** 2))
    print 'RMSE is ' + str(rmse)
    return rmse
```

### How does our RMSE change with different number of singular values?


```python
X_matrix, locators = create_test_matrix(r, 0.05)

preds = create_svd_predictions(r, 50, 1)
preds2 = create_svd_predictions(r, 50, demean=2)

get_rmse(r, preds, locators)
get_rmse(r, preds2, locators)
```

### Increasing the number of singular values decreases the RMSE.

```python
no_of_singular_values = []
rmse_list = []
rmse_list2 = []
for i in np.arange(10,150, 10):
    np.random.seed(42)
    X_matrix, locators = create_test_matrix(r, 0.2)
    preds = create_svd_predictions(r, i, 1)
    preds2 = create_svd_predictions(r, i, demean=2)
    
    print 'For ' +str(i) + ' number of singular values, our rmse is: '
    rmse_for_list = get_rmse(r, preds, locators)
    rmse_for_list2 = get_rmse(r, preds2, locators)
    no_of_singular_values.append(i)
    
    rmse_list.append(rmse_for_list)
    rmse_list2.append(rmse_for_list2)
        
tracker = pd.DataFrame({'Number of singular values':no_of_singular_values, 'RMSE product demean':rmse_list,
                       'RMSE user demean':rmse_list2})

tracker['Change in RMSE'] = tracker['RMSE product demean'].diff()
tracker['Change in RMSE 2'] = tracker['RMSE user demean'].diff()

tracker['Change in RMSE 2'] = tracker['RMSE user demean'].diff()

tracker.plot(x='Number of singular values', y=['RMSE user demean', 'RMSE product demean'], figsize=(9,9))
```

### Problem: What to do when a new user enters the dataset? It is computationally expensive to calculate SVD, we cannot possibly create a new prediction matrix each time a user enters.

### Solution: Employ a method called "folding in"

```python
from numpy.linalg import inv

U, sigma, Vt = svds(r_df, k = 10)
sigma = np.diag(sigma)
inversed_sigma = inv(sigma)
V = np.transpose(Vt)
test_user_vector = np.array(r_df.sample())
np.dot(np.dot(np.array(test_user_vector), V), inversed_sigma)
```

### Which method should I use?



* Jacard similarity: Not accurate, as it doesn't take into account the actual ratings. Under jacard similarity, ppl who have rated a lot of the same shows would be grouped as similiar, even if they have vastly different ratings for the shows.

* Cosine distance:  More towards content based filtering. Problem: cosine similarity treats missing rating as negative. Solution: use centred cosine (AKA Pearson correlation). Normalise ratings for a given user by subtracting his average rating. 

* User-user collaborative filtering: Make a set of users who have similar preferences to our target user. Use the ratings of our set of users to predict if our target user will like the movie.

* Item-item collaborative filtering: Instead of starting out with a user and finding similar user, we start with an item i, and find similar items to item i. Then, estimate the rating for item i based on ratings for similar items.
